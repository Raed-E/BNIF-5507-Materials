```python
#allow access to modules and import them
import sys
import os
sys.path.append(os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

from students.dimensionality_reduction import apply_pca, apply_tsne, apply_umap
from students.clustering import run_kmeans, run_hierarchical
from students.visualization import plot_representation_comparison, plot_cluster_comparison

#load the dataset
data = load_breast_cancer()

X = data.data
y = data.target

print(X.shape)
```

```python
#Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)
```

```python
#Dimensionality reduction
pca_result = apply_pca(X_scaled, n_components=2)
tsne_result = apply_tsne(X_scaled, n_components=2)
umap_result = apply_umap(X_scaled, n_components=2)

print(pca_result.shape)
print(tsne_result.shape)
print(umap_result.shape)
```

```python
#Visualizations
plot_representation_comparison(
    pca=pca_result,
    tsne=tsne_result,
    umap=umap_result,
    true_labels=y,
    filename="Figure1"
)
```

```python
#Clustering
n_clusters = 2
cluster_labels, cluster_centers = run_kmeans(X_scaled, n_clusters=n_clusters)
```

```python
#Visualize Clusters
plot_cluster_comparison(
    pca=pca_result,
    tsne=tsne_result,
    umap=umap_result,
    cluster_labels=cluster_labels,
    filename="Figure2"
)
```
